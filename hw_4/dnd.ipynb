```python
from matplotlib import pyplot as plt
import numpy as np
```

Lets define a python function to roll our die

```python
def roll_die_avg(num_rolls, num_sides):
    rolls =  np.random.randint(1, num_sides + 1, num_rolls)
    plt.hist(rolls, bins=np.arange(0.5, num_sides + 1.5, 1.0))

    return np.average(rolls)
```

Lets try 10 rolls of a 20 sided die

```python
roll_die_avg(10, 20)
```

Now lets try 1,000,000 rolls of a 20 sided die

```python
roll_die_avg(1000000, 20)
```

The average value of 10.5 is what I would expect, since a roll of zero is impossible the average should be slightly higher than the intuitive average of 10.

The histogram shows a fairly uniform distribution, which is to be expected since each side has an equal chance of being rolled.

Now lets write a function to roll a n sided die m times and returns the maximum value.

```python
def roll_die_max(num_rolls, num_sides):
    rolls =  np.random.randint(1, num_sides + 1, num_rolls)

    return np.max(rolls)
```

Now lets simulate rolling the die twice, and taking the max, repeating 1,000,000 times.

I'm having some fun here, and trying to do this without using any kind of loop to try to get better at using numpy efficiently

```python

# Creating a 2xnum_rolls array of random numbers between 1 and num_sides
def create_random(num_rolls, num_sides):
    return(np.random.randint(1, num_sides + 1, (num_rolls, 2)))


# Applying np.max over the 2nd axis (axis=1) of the array to get a single array containing the max of each row
max_of_two = np.apply_over_axes(np.max, create_random(1000000, 20), [1])

# Plotting the histogram of the max of each row
plt.hist(max_of_two, bins=np.arange(0.5, 20.5, 1.0))
```

This shape does make sense, as when youre always picking the largest of two rolls, you would need 2 low rolls to get a low value, whereas only one large roll to get a higher value.
